```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import pprint
import datetime as dt
from config import TM_API_Key
```

```python
pp = pprint.PrettyPrinter(indent=4)

with open('Concerts.json', 'r') as myfile:
    Lizzo_Data_Read=myfile.read()

# parse file
Lizzo_Data = json.loads(Lizzo_Data_Read)

# show values
embedded=Lizzo_Data["_embedded"]["events"][0]["_embedded"]["venues"][0]["city"]["name"]
events = Lizzo_Data["_embedded"]["events"]
```

```python
events_count = list(range(len(events)))
```

```python
cities = []
venue_ids = []
zip_codes = []
lats = []
lngs = []
event_dates = []
genre_ids = []
genre_names = []
max_prices = []
#days_of_the_week = []

for city in events_count:
    city_name = events[city]["_embedded"]["venues"][0]["city"]["name"]
    cities.append(city_name)  
    
    venue_id_name = events[city]["_embedded"]["venues"][0]["id"]
    venue_ids.append(venue_id_name)
    
    zip_code = events[city]["_embedded"]["venues"][0]["postalCode"]
    zip_codes.append(zip_code)
    
    lat = events[city]["_embedded"]["venues"][0]["location"]["latitude"]
    lats.append(lat)
    
    lng = events[city]["_embedded"]["venues"][0]["location"]["longitude"]
    lngs.append(lng)
    
    event_date = events[city]["dates"]["start"]["localDate"]
    event_dates.append(event_date)
    
    #day_of_the_week = event_date.day_name()
    #days_of_the_week.append(day_of_the_week)
    
    genre_id = events[city]["classifications"][0]["genre"]["id"]
    genre_ids.append(genre_id)

    genre_name = events[city]["classifications"][0]["genre"]["name"]
    genre_names.append(genre_name)
    
    try:
        max_price = events[city]["priceRanges"][0]["max"]
        max_prices.append(max_price)
    except:
        max_prices.append(0)
lngs
```

```python
url = "https://app.ticketmaster.com/discovery/v2/venues/"
#https://app.ticketmaster.com/discovery/v2/venues/KovZpZAFnIEA.json?apikey=edjeS6DGpS6D19CGdt9Pc7bdYR6Ab5yp"
```

```python
for venue_id in venue_ids:
    query_url = f'{url}{venue_id}.json?apikey={TM_API_Key}'
    response = requests.get(query_url).json()
    pp.pprint(response)
```

```python
Lizzo_Concerts_df = pd.DataFrame({
    "Venue ID": venue_ids,
    "City": cities,
    "Zip Code": zip_codes,
    "Latitude": lats,
    "Longitude": lngs,
    #"Day of The Week": days_of_the_week,
    "Date": event_dates,
    "Genre Name": genre_names,
    "Genre ID": genre_ids,
    "Max Ticket Price": max_prices
})
Lizzo_Concerts_df
```
